```python
import pandas as pd
cia = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/CIACountries.csv")
print(cia.columns)
cia.head

```

```python
cia.describe()
```

# Rename

```python
#I'm going to change the name of pop because that's a function in python.
cia = cia.rename(columns={"pop": "population"})
cia.head
```

```python
print(cia.shape)
print(type(cia.country))
```

# Filter and select together

```python
#Subset of columns 
cia[cia.population > 1000000000][["country","population","gdp"]]

```

# Mutate and rename

```python
cia["dens"] = cia["population"]/cia["area"]
cia.head()
```

# Sorting

```python
cia.sort_values(by = "population", ascending = False)
```

```python
cia.sort_values(by = "population", ascending = True)
```

# Summarize and grouping

```python
cia["high_educ"] = cia["educ"] > 4.5
cia[["high_educ","area"]].dropna(how = "any").groupby("high_educ").mean()


```

```python
cia.loc[cia["educ"].notna()].groupby("high_educ").mean()
```

```python
cia.loc[cia["educ"].notna()].groupby("high_educ").mean()["area"]
```
